### 2 - Exploring Dataset

```python
#see the total values of contain team_id
nba["team_id"].value_counts()
```

```python
#see the total values of contain fran_id
nba["fran_id"].value_counts()
```

```python
# see all the team for one franchise_id
nba.loc[nba["fran_id"] == "Lakers", "team_id"].value_counts()
```

```python
#define a colum to converts the value of date-game to the datetime data type
nba["date_played"] = pd.to_datetime(nba["date_game"])

#define a temporality with the first date and last date
nba.loc[nba["team_id"] == "MNL", "date_played"].min()
nba.loc[nba["team_id"] == "MNL", "date_played"].max()

#aggrete the new values
nba.loc[nba["team_id"] == "MNL", "date_played"].agg(("min", "max"))
```

```python
nba.loc[nba["fran_id"] == "Celtics", "team_id"].value_counts()
```

```python
nba.loc[nba["team_id"] == "BOS", "fran_id"].value_counts()
```

```python
# make the sum of total values to column pts
nba.loc[nba["team_id"] == "BOS", "pts"].sum()
```
